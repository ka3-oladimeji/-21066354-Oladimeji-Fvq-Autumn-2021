### Requirement FR9 - Read CSV data from two files and merge it into a single Data Frame 

```python
import pandas as pd
from IPython.display import display

# 'pd.task2a.csv'
# 'pd.task2b.csv


task2a_df = pd.read_csv('task2a.csv',encoding= 'latin')
task2b_df = pd.read_csv("task2b.csv",encoding= 'latin')

task2_df = task2a_df.merge(task2b_df, on='region')

task2_df.to_csv('task2_merge.csv', index=False)

display(task2_df)

```

### Requirement FR10 - Clean the merged data

```python
import pandas as pd
from IPython.display import display

# 'pd.task2a.csv'
# 'pd.task2b.csv


task2_df = pd.read_csv('task2_merge.csv',encoding= 'latin')

# Remove rows containing missing values
df_removed = task2_df.dropna(axis=0, how='any')

df_cleaned = df_removed.drop('share_voters_voted_to_the_right', axis=1)

df_cleaned.to_csv('task2_clean.csv', index=False)

print('After removing rows')
print(df_removed)

print('After removing share_voters_voted_to_the_right column')
print(df_cleaned)

```

### Requirement FR11 - Investigate the effects of household income on levels of bias crimes

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

task2_df = pd.read_csv('task2_clean.csv',encoding= 'latin')

sb.scatterplot(x="median_household_income", y="avg_biascrimes_per_100k", data=task2_df)


plt.show()
```

### Requirement FR12 - Test the hypothesis that household income has no effect on levels of bias crime

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

task2_df = pd.read_csv('task2_clean.csv',encoding= 'latin')

stat, p = pearsonr(task2_df['median_household_income'], task2_df['avg_biascrimes_per_100k'])

print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```
